```python
import pandas as pd
import numpy as np
import requests
import json
import os
from bs4 import BeautifulSoup
import collections
collections.Callable = collections.abc.Callable
import time
```

```python
geniustoken = os.environ['geniustoken']
useragent_url = 'https://httpbin.org/user-agent'
r = requests.get(useragent_url)
useragent = json.loads(r.text)['user-agent']
```

```python
urltoscrape = 'https://genius.com/Lizzo-good-as-hell-lyrics'

r = requests.get(urltoscrape,
                headers = {'User-Agent': useragent,
                          'From': 'tby8aj@virginia.edu'})
myhtml = BeautifulSoup(r.text,'html.parser')
r
```

```python
# (a) Find the lyrics to "Good as Hell"

myhtml.find_all('div', attrs={'data-lyrics-container':True})[0].text
```

```python
# (b) Find links to the other songs and append to song_list

song_list=[]

for x in range(4,19):
    song_list.append(myhtml.find_all('a', attrs={'class':"Link__StyledLink-h3isu4-0 fUgcxf"})[x])

song_list
```

```python
song_links=[]

for x in song_list:
    song_links.append(x.get('href'))
    
song_links
```

```python
def scrape_song(url):
    time.sleep(5)
    print('Now getting the lyrics from ' + url)
    r = requests.get(url, headers = {'User-Agent': useragent, 'From': 'tby8aj@virginia.edu'})
    myhtml = BeautifulSoup(r.text,'html.parser')
    try:
        lyrictext = myhtml.find_all('div', attrs={'data-lyrics-container':True})
        return lyrictext
    except: pass
```

```python
lyrics = [scrape_song(u) for u in song_links]
```

```python
headers = {'User-Agent': useragent,
          'From': 'tby8aj@virginia.edu'}

root = 'https://maps.spirithalloween.com'

endpoint = '/api/getAsyncLocations'

myparams = {'template':'domain','level':'domain','lat':40.380028, 'lng':-97.910156, 'radius':1800, 'limit':1500}

r = requests.get(root + endpoint,
                headers = headers,
                params = myparams)

myjson = json.loads(r.text)
spirit = pd.json_normalize(myjson, record_path = ['markers'])
spirit
```

```python
mylist = [BeautifulSoup(x).find_all('div', attrs={'class':'address-two'})[0].text for x in spirit['info']]
mylist
```
